```python
import os
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv('heart.csv')
df.head()
```

```python
sns.displot(df['Age'], kde = True, color = 'blue')
```

```python
sns.displot(x = df['RestingBP'], kde = True, color = 'blue')
```

```python
sns.displot(df['Cholesterol'], kde = True, color = 'blue')
```

```python
sns.displot(df['MaxHR'], kde = True, color = 'blue')
```

Categorical Columns (Pie Charts)

```python
df.groupby('Sex').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

```python
df.groupby('ChestPainType').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

```python
df.groupby('RestingECG').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

```python
df.groupby('ST_Slope').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

```python
df.groupby('HeartDisease').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

ViolinPlot.

```python
sns.violinplot(x = df['Age'])
```

```python
sns.violinplot(y = df['Sex'], x = df['HeartDisease'])
```

```python
sns.violinplot(y = df['Age'], x = df['HeartDisease'])
```

```python
sns.violinplot(y = df['Cholesterol'], x = df['HeartDisease'])
```

Correlation (HeatMap)

```python
df.corr(numeric_only=True)
# The numeric_only=True argument tells the corr() method to only
# include numeric columns in the correlation calculation.
```

```python
sns.heatmap(df.corr(numeric_only=True)) # The numeric_only=True argument tells the corr() method to only include numeric columns in the correlation calculation.
```

Correlation (PairPlot)

```python
sns.pairplot(df)
```

Correlation (JointPlot)

```python
sns.jointplot(x = 'Age' , y = 'MaxHR', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'Age' , y = 'MaxHR', data = df, kind = 'reg')
```

```python
sns.jointplot(x = 'Cholesterol' , y = 'MaxHR', data = df, kind = 'reg')
```
